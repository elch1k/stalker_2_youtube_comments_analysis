```python
# get video_id from each url
video_ids = []

for url in video_urls:
  video_id = url[-11:]
  video_ids.append(video_id)
```

## Create pivot table of videos

```python
df = pd.DataFrame({
    "video_title": [],
    "video_id": [],
    "published_at": [],
    "channel_name": [],
    "views_count": [],
    "like_count": [],
    "comment_count": [],
})

api_service_name = "youtube"
api_version = "v3"

youtube = googleapiclient.discovery.build(
    api_service_name, api_version, developerKey=API_KEY)

for video_id in video_ids:
  request = youtube.videos().list(
      part="snippet,statistics,contentDetails",
      id=video_id
  )
  response = request.execute()

  item = response["items"][0]
  video_title = item.get("snippet").get("title")
  channel_name = item.get("snippet").get("channelTitle")
  published_at = item.get("snippet").get("publishedAt")
  views_count = item.get("statistics").get("viewCount")
  like_count = item.get("statistics").get("likeCount")
  comment_count = item.get("statistics").get("commentCount")

  temp_frame = pd.DataFrame({
    "video_title": [video_title],
    "video_id": [video_id],
    "published_at": [published_at],
    "channel_name": [channel_name],
    "views_count": [views_count],
    "like_count": [like_count],
    "comment_count": [comment_count],
    })

  df = pd.concat([df, temp_frame], ignore_index=True)
```

```python
# add id column
df.reset_index(inplace=True)
df.rename(columns={"index": "id"}, inplace=True)
```

## Comment collection from video

Here I scrape all main comments below each video. Main means without reply.

```python
def get_youtube_video_comment(uniq_id: str, video_id: str, API_KEY: str):
  comments = []
  api_service_name = "youtube"
  api_version = "v3"

  youtube = googleapiclient.discovery.build(
    api_service_name, api_version, developerKey=API_KEY)

  request = youtube.commentThreads().list(
      part="snippet",
      videoId=video_id,
      maxResults=100  # 100 is possible max
  )

  while request:
    # execute the request
    response = request.execute()
    # get the comments data from the response
    for item in response["items"]:
        comment = item["snippet"]["topLevelComment"]["snippet"]
        reply_count = item["snippet"]["totalReplyCount"]
        public = item["snippet"]["isPublic"]
        comments.append([
            uniq_id,
            comment["authorDisplayName"],
            comment["publishedAt"],
            comment["likeCount"],
            comment["textOriginal"],
            reply_count,
            public
        ])
    if response.get("nextPageToken"):
      # overwrite a new request object with the next page token
      request = youtube.commentThreads().list(part="snippet",
                                              videoId=video_id,
                                              maxResults=100,
                                              pageToken=response.get("nextPageToken"))
    else:
      break

  df = pd.DataFrame(comments, columns=["id", "author", "published_at",
                                       "like_count", "text", "reply_count", "public"])
  return df
```

```python
comment_df = pd.DataFrame({
    "id": [],
    "author": [],
    "published_at": [],
    "like_count": [],
    "text": [],
    "reply_count": [],
    "public": [],
})

for i in df.itertuples():
  uniq_id = i[1]
  vid_id = i[3]
  temp_data = get_youtube_video_comment(uniq_id=uniq_id,
                                        video_id=vid_id,
                                        API_KEY=API_KEY)
  comment_df = pd.concat([comment_df, temp_data], ignore_index=True)
```
